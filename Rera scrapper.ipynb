```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from webdriver_manager.firefox import GeckoDriverManager
import time
from bs4 import BeautifulSoup
import requests as req
import regex as re
from tqdm import tqdm

from selenium.webdriver.firefox.options import Options
options = Options()
options.headless = False

driver = webdriver.Firefox(executable_path=GeckoDriverManager().install(), options=options) #'/home/sanket/chromedriver_linux64/chromedriver'
```

```python
# Goes to the base url
driver.get("https://maharerait.mahaonline.gov.in/")
```

```python
# Selects select project details button on homepage
search_project_details_button = driver.find_element_by_xpath('/html/body/section/div/div/form/div[1]/div[3]/div/div/div[2]/div[2]/a')
search_project_details_button.click()
```

```python
# Select Registered projects
driver.find_element_by_xpath('//*[@id="Promoter"]').click()
```

```python
# select_advanced_search
driver.find_element_by_xpath('//*[@id="btnAdvance"]').click()
```

```python
# selects state 
select_state = driver.find_element_by_xpath('//*[@id="State"]')
which_state = input('Please Enter State: ')
for option in select_state.find_elements_by_tag_name('option'):
    if option.text == which_state.upper():
        option.click() 
        break
```

```python
# Selects district
select_district = driver.find_element_by_xpath('//*[@id="District"]')
which_district = input('Please Enter district: ')
for option in select_district.find_elements_by_tag_name('option'):
    if option.text == which_district:
        option.click() 
        break
```

```python
# Start the search
search_button = driver.find_element_by_xpath('//*[@id="btnSearch"]')
search_button.click()
```

```python
%%time

view_links = [] # Create an accumulator

total_pages = driver.find_element_by_id('TotalPages').get_attribute('value')

for i in tqdm(range(int(total_pages) - 1)):
    all_rows = driver.find_elements_by_css_selector("td:nth-child(5) > b:nth-child(1) > a:nth-child(1)")
    for row in all_rows:
        view_links.append(row.get_attribute('href')) 
        
    time.sleep(1)
    next_button = driver.find_element_by_xpath('//*[@id="btnNext"]')
    next_button.click()
```
